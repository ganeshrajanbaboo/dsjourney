# Exploration Cuantitative Features 

**The soil type categorical features can be combined as one feature as only one type is true for one observation**

```python
#For each column of the input row, check which column has value 1 and return the name of that column
def mergecols(row, soil_cols):
    for col in soil_cols:
        if row[str(col)] == 1:
            return col

#Merge all columns that start "Soil_Type" into one column with value in each row as the column returned by mergecols
#Further remove the prefix, "Soil_Type"
def merge_action(df):
    soil_cols = [col for col in df if col.startswith('Soil_Type')]
    df['Soil_Category'] = df.apply(mergecols, soil_cols=soil_cols,axis=1)
    df['Soil_Category'] = df.Soil_Category.map(lambda x: x.strip('Soil_Type')) #remove the "Soil_Type" prefix
    df.Soil_Category = df.Soil_Category.astype("float")
    df.drop(columns=soil_cols, inplace=True)

merge_action(X_train)
merge_action(X_validate)

X_train.loc[:, 'Soil_Category'] = X_train.loc[:, 'Soil_Category'].astype("category")
X_validate.loc[:, 'Soil_Category'] = X_validate.loc[:, 'Soil_Category'].astype("category")
```

### Some features look near normal, some skewed and some with multiple peaks. Let's try some normalization using some feature composition.

### Import some python libraries

```python
from scipy import stats
from scipy.stats import boxcox
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import normaltest
```

### The parameters of this feature composition exercise were arrvied at by trial and error until a best possible fit was obtained that near normalize the resultant feature measurable by the qqplot and the p-value of normaltest
### The step was tuned until p-value obtained was > 0.4; so considering an alpha = 0.05 we can safely fail to reject the "normalness"

```python
def feature_comp_1(df):
    resultant_1 = df['Hillshade_9am']/100 + df['Hillshade_Noon'] + df['Hillshade_3pm'] + df['Horizontal_Distance_To_Roadways']/30
    resultant_1 = resultant_1 - df['Slope']
    resultant_1 = boxcox(resultant_1, 1.5) # this has the effect of raising to the power 1.5
    return resultant_1

rt = feature_comp_1(X_train)
X_train['resultant_1'] = pd.Series(rt)
rv = feature_comp_1(X_validate)
X_validate['resultant_1'] = pd.Series(rv)

f = plt.figure(figsize=(15,7))
ax = f.add_subplot(1, 2, 1)
sns.distplot(X_train['resultant_1'])
ax = f.add_subplot(1, 2, 2)
qqplot(X_train['resultant_1'], line='s', ax=ax)
stat, p = normaltest(X_train['resultant_1'])
print("p-value = %.5f" %p)
```

### Lets's see if the p-value of "normalness" holds when for individual classes of the label.
### Hmm.. looks not. Nevertheless, let's go ahead

```python
f = plt.figure(figsize=(20, 9))
for i in range(1,8):
    ax = f.add_subplot(2, 4, i)
    ax.set_title = "class = " + str(i) #for some reason the title wouldn't appear on the graphs :( Let me know if you can tell. Thanks.
    sns.distplot(X_train.loc[y_train == i].resultant_1)
    stat, p = normaltest(X_train.loc[y_train == i].resultant_1)
    print('"normalness" p-value for class %.0f is %.30f' %(i, p))
```

### Remove the columns that were feature engineered on to produce resultant_1

### A random forest model would required that there is minimum correlation between the features. Let us check how the correlation looks

```python
X_train.corr()
```

### We see a strong correlation between 'Horizontal_Distance_To_Fire_Points' and 'Vertical_Distance_To_Hydrology'. Let's see what we can do about it.

### The feature composition couldn't take it any near to normalness (p_value exponentially less than 0.05), but we'll buy it.

```python
def feature_comp_2(df):
    hydrology_dist = df['Horizontal_Distance_To_Hydrology'] - df['Vertical_Distance_To_Hydrology'] * 100
    hydrology_dist += 100000
    hydrology_dist = boxcox((hydrology_dist/5), 0)
    hydrology_dist = boxcox(hydrology_dist, 73)
    return hydrology_dist

X_train['hydrology_dist'] = pd.Series(feature_comp_2(X_train))
X_train.drop(['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology'], axis=1, inplace=True)

X_validate['hydrology_dist'] = pd.Series(feature_comp_2(X_validate))
X_validate.drop(['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology'], axis=1, inplace=True)

sns.distplot(X_train['hydrology_dist'])
stat, p = normaltest(X_train['hydrology_dist'])
print(p)
```

### Scale continuous variables using scikit learn's standard scaler

```python
from sklearn.preprocessing import StandardScaler
def scale_contfeatures(df):
    contvar_cols = ['Elevation', 'Aspect', 'Horizontal_Distance_To_Fire_Points', 'resultant_1', 'hydrology_dist']
    df_scale = df[contvar_cols]
    X_scaler = StandardScaler()
    df_scale = X_scaler.fit_transform(df_scale)
    for i, col in enumerate(contvar_cols):
        df[col] = pd.Series(df_scale[:, i])
    
scale_contfeatures(X_train)
scale_contfeatures(X_validate)
```
